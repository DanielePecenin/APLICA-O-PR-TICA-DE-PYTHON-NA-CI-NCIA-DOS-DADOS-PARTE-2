```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_excel('C:\\Users\\danip\\Downloads\\default_of_credit_card_clients__courseware_version_1_21_19.xls')
```

```python
# Definir as listas de colunas para valores da fatura e pagamentos
bill_amt_columns = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
pay_amt_columns = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

# Exibir as listas
print("Bill Amount Columns:", bill_amt_columns)
print("Pay Amount Columns:", pay_amt_columns)
```

```python
# Exercício 2: Analisar as sínteses estatísticas dos valores da fatura
print("Estatísticas descritivas para os valores da fatura:")
print(df[bill_amt_columns].describe())
```

```python
# Exercicio 3 - Visualizar as características de valor da fatura com histogramas
plt.figure(figsize=(12, 8))

# Grade 2x3 de plotagens de histograma
for i, col in enumerate(bill_amt_columns, 1):
    plt.subplot(2, 3, i)
    plt.hist(df[col].dropna(), bins=20, edgecolor='black')
    plt.title(f'Histograma de {col}')
    plt.xlabel('Valor da fatura')
    plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
# Exercicio 4 - Obter resumo das características de valor do pagamento
print("Estatísticas descritivas para os valores do pagamento:")
print(df[pay_amt_columns].describe())
```

```python
# Exercicio 5 - Plote um histograma para as características de pagamento da fatura com rotação nos rótulos
plt.figure(figsize=(12, 8))

for i, col in enumerate(pay_amt_columns, 1):
    plt.subplot(2, 3, i)
    plt.hist(df[col].dropna(), bins=20, edgecolor='black')
    plt.title(f'Histograma de {col}')
    plt.xlabel('Valor do pagamento')
    plt.ylabel('Frequência')
    plt.xticks(rotation=45)  # Rotacionando os rótulos do eixo X

plt.tight_layout()
plt.show()
```

```python
# Exercicio 6 - Verifique quantos dados de valor de pagamento são exatamente iguais a 0
zero_payments = df[pay_amt_columns].apply(lambda x: x == 0).sum()
print("\nQuantos pagamentos são exatamente iguais a 0?")
print(zero_payments)
```

```python
# Exercicio 7 - Plotar histogramas de transformações logarítmicas dos pagamentos diferentes de zero
plt.figure(figsize=(12, 8))

# Aplicar log10 nos valores de pagamento que são diferentes de zero
for i, col in enumerate(pay_amt_columns, 1):
    plt.subplot(2, 3, i)
    log_values = np.log10(df[col].replace(0, np.nan).dropna())  # Substituir 0 por NaN e aplicar log10
    plt.hist(log_values, bins=20, edgecolor='black')
    plt.title(f'Log10 de {col} (pagamentos > 0)')
    plt.xlabel('Log10 do valor do pagamento')
    plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
